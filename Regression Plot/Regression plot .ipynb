```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("scatter_plot_ii.csv")
data.head(5)
```

## sns.regplot

```python
#plt.figure(figsize=(10,8))
sns.set(rc={
    'figure.figsize': (9,6)
})
sns.regplot(x="Budget",
            y="Sales",
            data=data,
            scatter_kws = {'color':'midnightblue'},
            line_kws = {'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, weight = 'bold' )
plt.show()
```

## sns.lmplot

```python
sns.lmplot(x= "Budget",
           y= "Sales",
           data= data,
           height = 8,
           scatter_kws = {'color':'midnightblue'},
           line_kws = {'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, weight = 'bold' )
plt.show()
```
