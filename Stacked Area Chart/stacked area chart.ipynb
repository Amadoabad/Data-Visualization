```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
sns.set()
```

```python
data = pd.read_csv("stacked_area_chart_data.csv")
```

```python
data.head(2)
```

```python
colors = ['#011638', '#7E2987', "#EF2026"]
labels =["Diesel", "Petrol", "Gas"]
sns.set_style("dark")
plt.figure(figsize=(12,6))
plt.stackplot(data["Year"]
             ,data["Diesel"]
             ,data["Petrol"]
             ,data["Gas"]
             ,colors = colors
             ,edgecolor = "none")
plt.xticks(data["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of Cars", fontsize =13 )
plt.title("Popularity of Engine Fuel Types (1962 - 2016)"
          ,fontsize = 14, weight = "bold" )
#sns.despine()
plt.show()
```
