```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import plotly.express as px
```

```python
data = pd.read_csv("line_chart_data.csv")
data.head(2)
```

```python
data["new_date"] = pd.to_datetime(data["Date"])
data["new_date"]
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(data["new_date"], data["GSPC500"])
plt.plot(data["new_date"], data["FTSE100"])
plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize = 14, fontweight ="bold")
plt.xlabel("Date")
plt.ylabel("Returns")
plt.legend(labels = labels, fontsize ="large" )
plt.show()
```

```python
data_h2_08 = data[(data["new_date"] >= "2008-07-01") &
                  (data["new_date"] <= "2008-12-31")]
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(data_h2_08["new_date"], data_h2_08["GSPC500"], color ="midnightblue")
plt.plot(data_h2_08["new_date"], data_h2_08["FTSE100"], color ="crimson")
plt.title("S&P vs FTSE Returns (H2 2008)", fontsize = 14, fontweight ="bold")
plt.xlabel("Date")
plt.ylabel("Returns")
plt.legend(labels = labels, fontsize ="large" )
plt.show()
```
