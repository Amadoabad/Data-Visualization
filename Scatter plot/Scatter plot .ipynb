```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("scatter_data.csv")
data.head(5)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,8))
scatter =plt.scatter(data["Area (ft.)"],
            data["Price"],
            alpha=0.6, 
            c = data["Building Type"],
            cmap= 'viridis')
plt.legend(*scatter.legend_elements(),
           loc = "upper left",
           title = "Building Type")
plt.title("Realtionship between Area and Price for Real Estates.", fontsize=14, weight= "bold")
plt.xlabel("Area (sq.ft.)", weight = "bold")
plt.ylabel("Price (000' $)", weight = "bold")

plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x= data["Area (ft.)"], y=data["Price"],
                hue = data["Building Type"], palette = ["black", "darkblue", "purple", "pink", "white"],
                s = 100)
plt.title("Realtionship between Area and Price for Real Estates.", fontsize=14, weight= "bold")
plt.xlabel("Area (sq.ft.)", weight = "bold")
plt.ylabel("Price (000' $)", weight = "bold")

plt.show()
```
